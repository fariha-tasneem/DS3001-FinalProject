## Methods - KNN Model Preprocessing

```python
# Collapse the NObeyesdad variable 
underweight = ['Insufficient_Weight']
normal = ['Normal_Weight']
overweight = ['Overweight_Level_I', 'Overweight_Level_II']
obese = ['Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III']

# Function to categorize weight
def categorize_weight(row):
    if row in underweight:
        return 'underweight'
    elif row in normal:
        return 'normal'
    elif row in overweight:
        return 'overweight'
    elif row in obese:
        return 'obese'
    
# Apply the function to create a new column
obesity['weight_category'] = obesity['NObeyesdad'].apply(categorize_weight)

obesity
obesity.weight_category.value_counts()
```

## Methods - Model

```python
# Check the prevalence of the target class
obesity['weight_category_overweight'].value_counts()[1] / obesity['weight_category_overweight'].count()
```

```python
X = obesity[["family_history_with_overweight_yes","FAVC_yes","FCVC", "FAF", "TUE"]]
y = obesity['weight_category_overweight'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80, stratify = y, random_state=21)
```

```python
X_tune, X_test, y_tune, y_test = train_test_split(X_test,y_test,  train_size = 0.50, stratify = y_test,random_state=49)
```

```python
obesity_3NN = KNeighborsClassifier(n_neighbors=9)
```

```python
obesity_3NN.fit(X_train, y_train)
```

```python
# Probability that our model predicted the correct output based on given inputs

obesity_3NN.score(X_tune, y_tune) 
```

## Evaluation of model


```python
obesity_prob1 = obesity_3NN.predict_proba(X_tune) #This function gives percent probability for both class (0,1)
obesity_prob1[:5] 
```

```python
obesity_pred1 = obesity_3NN.predict(X_tune)
obesity_pred1[:20]
```

```python
final_model= pd.DataFrame({'neg_prob':obesity_prob1[:, 0], 'pred':obesity_pred1,'target':y_tune, 'pos_prob':obesity_prob1[:, 1]})
final_model.head()
```

```python
metrics.confusion_matrix(final_model.target,final_model.pred)
```

```python
metrics.ConfusionMatrixDisplay.from_predictions(final_model.target,final_model.pred, display_labels = [False, True])
```

#### Based on the confusion matrix, the model is better at classifying negative classes (not obese) than positive classes (obese). This is possibly due to the fact that the amount of data for the obesity class is relatively small.

```python
# Function so we can make the threshold whatever we want, not just the default 50%
def adjust_thres(x,y,z):
  #x=pred_probablities, y=threshold, z=test_outcome
  thres = (np.where(x > y, 1,0))
  #np.where is essentially a condensed if else statement. The first argument is the condition, then the true output, then the false output
  return metrics.ConfusionMatrixDisplay.from_predictions(z,thres, display_labels = [False, True], colorbar=False)
```

```python
# With a threshold of .35
adjust_thres(final_model.pos_prob,.35,final_model.target)
```

```python
#Now let's use our model to obtain an ROC Curve and the AUC
metrics.RocCurveDisplay.from_predictions(final_model.target, final_model.pos_prob)
#Set labels and midline...
plt.plot([0, 1], [0, 1],'r--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```

```python
# Let's extract the specific AUC value now
metrics.roc_auc_score(final_model.target, final_model.pos_prob) 
```

#### The AUC value of 0.74 indicates that the model has relatively good discriminative power in distinguishing between positive and negative instances (classifying individuals as obese or not obese).

```python
# Determine the log loss
metrics.log_loss(final_model.target, final_model.pos_prob)
```

#### The log loss value of 0.505966 means that, on average, the predicted probabilities generated by the model are relatively close to the true probabilities for obesity classification, but it definitely has room for improvement.

```python
# Get the F1 Score
metrics.f1_score(final_model.target, final_model.pred)
```

#### The F1 score of 0.4565217 indicates a moderate performance of the model in classifying obesity. 

```python
# Other evaluation metrics
print(metrics.classification_report(final_model.target, final_model.pred)) 
```

## Fairness assessment



Fairness and bias considerations are important to look at while building this model. One possible consideration is the lack of sample size with a little over 2k data points, which would raise concerns about the model's ability to generalize to broader populations and the risk of sampling error. However, since this data set is limited to the countries of Mexico, Peru, and Colombia, we believe that this model can still accurately reflect patterns among this population. While talking about the geography of this data, we must make sure to only limit generations of this model to this geographic region and not overextend the results to other regions, which would be biased. We must also consider how protected classes such as gender and age could skew impact fairness and that there is not an imbalance in the representation of the data. According to our EDA, there is a relatively equal number of females and males in the dataset, so there is no notable imbalance. Age is skewed towards age '20'. While this skew might impact the model's performance, further exploration can help mitigate potential biases arising from this age distribution imbalance.


Additionally, the dataset's potential for evaluation bias is heightened by its exclusivity to Mexico, Peru, and Colombia. This restriction might lead to an inaccurate assessment of the model's performance, especially if it is used in regions with different health characteristics or lifestyle factors. The limited representation could result in biased conclusions about the model's effectiveness in diverse settings. There is also the possibility of cognitive bias, particularly confirmation bias, is something to consider with this model. If there is an assumption that factors like family history strongly influence obesity without considering alternative explanations, the model may inadvertently learn and perpetuate these biases. Regardless, this model can be implemented as long as these bias and fairness considerations are kept in mind and assumptions are not overextended.
